### Mobile Phone Class (Easy)
#### üéØ Scenario:

Create a class `Mobile`.

**Requirements:**

-   Class variable: `brand_name = "Samsung"`

-   Instance variables:

    -   model

    -   price

-   Instance method:

    -   `show_details()` ‚Üí print model, price, brand

````python
class Mobile:
    # class varaible,
    brand_name = "Samsung"
    # The constructor (__init__) method initializes the attributes of the object
    def __init__(self, model, price):
        self.model = model  # Attribute to store the model of the mobile
        self.price = price  # Attribute to store the price of the mobile

    # Method to describe the mobile
    def show_details(self):
        print(f"The mobile is a {Mobile.brand_name} {self.model} and costs ${self.price}.")

# Create an object of the Mobile class with specific model and price
m1 = Mobile("A15", 50000)
# Call the show_details method to print details about the mobile
m1.show_details()
````

### Bank Account Class
#### üéØ Scenario:

Create a class BankAccount.

**Requirements:**

-   Instance variables:

    -   account_holder

    -   balance

-   Methods:

    -   deposit(amount)

    -   withdraw(amount)

    -   show_balance()

**Rules:**

-   Cannot withdraw more than balance

-   Cannot deposit negative amount

````python
class BankAccount:

    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self.balance = balance

    def deposit(self, amount):
        if amount <= 0:
            print("Deposit amount must be greater than 0.")
            return

        self.balance += amount
        print("Deposit Successful")
        print(f"{self.account_holder} deposited Rs.{amount}")
        print(f"New balance: Rs.{self.balance}")

    def withdraw(self, amount):
        if amount <= 0:
            print("Withdraw amount must be greater than 0.")
            return

        if amount > self.balance:
            print("Insufficient balance.")
            return

        self.balance -= amount
        print("Withdrawal Successful")
        print(f"{self.account_holder} withdrew Rs.{amount}")
        print(f"Remaining balance: Rs.{self.balance}")

    def show_balance(self):
        print(f"{self.account_holder}'s Balance: Rs.{self.balance}")


# Testing
account1 = BankAccount("Tariq Javed", 50000)

account1.deposit(10000)
account1.withdraw(5000)
account1.show_balance()
````
