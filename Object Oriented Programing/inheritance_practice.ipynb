## ðŸŸ¢ Practice 1 â€“ Single Inheritance (Easy)
### ðŸŽ¯ Scenario: Vehicle System
#### Parent Class â†’ `Vehicle`

**Attributes:**

-   brand
-   speed

**Method:**

-   `move()` â†’ prints: "Vehicle is moving at X km/h"

**Child Class â†’ `Car`**

Extra attribute:

-   fuel_type

Extra method:

-   `car_info()` â†’ prints brand, speed, fuel type

-   Use `super()` properly.

```python



class Vehicle:
    def __init__(self , brand , speed):
        self.brand = brand
        self.speed = speed

    def move(self):
        print(f"{self.brand} is moving at {self.speed} km/h.")


class Car(Vehicle):
    def __init__(self , brand , speed , fuel_type):
        super().__init__(brand , speed)
        self.fuel_type = fuel_type

    def car_info(self):
        print(f"Car Info â†’ Brand: {self.brand}, Speed: {self.speed} km/h, Fuel: {self.fuel_type}")



car =Car("Toyota" , 200 , "Petrol")
car.move()
car.car_info()
```

## ðŸŸ¡ Practice 2 â€“ Multilevel Inheritance

```python
class Person:
    def __init__(self, name):
        self.name = name

    def introduce(self):
        print(f"Hi, I am {self.name}")


class Employee(Person):
    def __init__(self, name, salary):
        super().__init__(name)
        self.salary = salary

    def show_salary(self):
        print(f"{self.name} earns Rs.{self.salary}")


class Manager(Employee):
    def __init__(self, name, salary, department):
        super().__init__(name, salary)
        self.department = department

    def show_department(self):
        print(f"{self.name} manages the {self.department} department")


# Testing
manager = Manager("Ali", 50000, "IT")

manager.introduce()        # From Person
manager.show_salary()      # From Employee
manager.show_department()  # From Manager
```

## ðŸŸ¢ Practice â€“ Multiple Inheritance

```python
class Father:

    def skills(self):
        print("Father: Gardening, Driving")

class Mother:

    def skills(self):
        print("Mother: Cooking, Painting")

class Child(Father , Mother):

    def show_all_skills(self):
        Father.skills(self)
        Mother.skills(self)


c = Child()
c.show_all_skills()


```

**Create Parent class â†’ Person**

-   name
-   age
-   method: `introduce()`

**Child class â†’ Student**

-   inherits Person

-   adds: grade

-   method: `show_grade()`

-   Use `super()` properly.

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        print(f"{self.name} is {self.age} years old.")


class Student(Person):
    def __init__(self, name, age, grade):
        super().__init__(name, age)
        self.grade = grade

    def show_grade(self):
        print(f"{self.name} has grade {self.grade}.")


# Testing
s1 = Student("Tariq", 15, "A")

s1.introduce()
s1.show_grade()
```
