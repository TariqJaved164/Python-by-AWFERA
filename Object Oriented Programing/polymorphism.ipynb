## Polymorphism

Polymorphism is a core concept in object-oriented programming (OOP). The word comes from Greek: **_poly_** (many) and **_morph_** (forms).

In programming, **polymorphism means the same interface can be used with different types of objects.**


In simple terms: you can write code that works on objects of multiple classes, as long as they support the operations you need. Python, being dynamically typed, makes this especially easy and natural.

## A Real‑World Analogy
Think of a universal remote control. It has a **"power"** button. You can point it at a TV, a stereo, or an air conditioner – each device understands "power" but does something different (TV turns on/off, stereo starts playing, AC starts cooling). The remote doesn't need to know exactly what kind of device it is; it just needs the device to respond to the "power" command.

## A Real‑World Analogy 2

Think of the word "Cut."

If you tell a hairstylist to "Cut," they grab scissors.

If you tell an actor to "Cut," they stop acting.

If you tell a chef to "Cut," they grab a knife.

The command is the same ("Cut"), but the action changes based on who is doing it. That is polymorphism!


In Python, polymorphism works the same way: different objects can respond to the same method call in their own way.

## Why Is Polymorphism Useful?

-   **Code reusability** – Write generic code that works with many object types.

-   **Flexibility** – Easily add new classes without changing existing code.

-   **Maintainability** – Changes in one class don't ripple through unrelated code.

## 1. Duck Typing
Python relies on **duck typing:** "**_If it walks like a duck and quacks like a duck, then it's a duck."_** This means we don't care about the actual type of an object; we only care whether it has the methods or attributes we need.

Example: Different animals, same method callm

```python
class Duck:
    def sound(self):
        return "Quack!"

class Dog:
    def sound(self):
        return "Woof!"

class Cat:
    def sound(self):
        return "Meow!"

# A function that uses duck typing
def make_sound(animal):
    print(animal.sound())

# All three classes have a sound() method, so they work!
make_sound(Duck())   # Quack!
make_sound(Dog())    # Woof!
make_sound(Cat())    # Meow!
```

Here, `make_sound()` doesn't check the class of `animal`. It just calls `.sound()`. Any object that provides `sound()` can be used.

## 2. Method Overriding (Polymorphism with Inheritance)
When a child class provides its own version of a method that is already defined in its parent class, we call it **method overriding**. The same method name behaves differently based on the object's actual class.

Example: Shape area calculation

```python
class Shape:
    def area(self):
        return "Unknown area"

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2

class Triangle(Shape):
    def __init__(self , base , height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base *self.height

# A list of different shapes
shapes = shapes = [Rectangle(4 , 5) , Triangle(4 , 5) , Circle(3)]

# Polymorphic behaviour: each object's area() is called
for shape in shapes:
    print(shape.area())   # Output: 20, 28.26
```

## 3. Operator Overloading
Python allows you to define how operators (like `+, -, *, ==`) behave for your own classes. This is a form of polymorphism because the same operator works differently depending on the objects involved.

Example: Adding two custom vectors

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        # Define what '+' means for two Vector objects
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

v1 = Vector(2, 3)
v2 = Vector(4, 5)
v3 = v1 + v2          # Calls v1.__add__(v2)
print(v3)             # Vector(6, 8)
```

## 4. Polymorphism with Functions and Objects
Because Python is dynamically typed, a single function can handle objects of completely different types as long as they support the required operations.

Example: A function that calculates total area of any objects with an `area()` method

```python
class Shape:
    def area(self):
        return "Unknown area"

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2

class Triangle(Shape):
    def __init__(self , base , height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base *self.height

# A list of different shapes
shapes = [Rectangle(4 , 5) , Triangle(4 , 5) , Circle(3)]

# Polymorphic behaviour: each object's area() is called
for shape in shapes:
    print(shape.area())   # Output: 20, 10.0 , 28.26


def total_area(objects):
    total = 0
    for obj in objects:
        total += obj.area()
    return total

print(f"{total_area(shapes):.2f}")
```

We didn't need to change `total_area()` when we added a new shape; it just works.

## 5. Abstract Base Classes (Enforcing an Interface)
Sometimes you want to guarantee that certain methods exist. Python's `abc` module lets you create abstract base classes that define a required interface. This is a more formal way of ensuring polymorphism.

Example: Using `ABC` to define a `Drawable` interface

```python
from abc import ABC, abstractmethod

class Drawable(ABC):
    @abstractmethod
    def draw(self):
        pass

class Circle(Drawable):
    def draw(self):
        print("Drawing a circle")

class Square(Drawable):
    def draw(self):
        print("Drawing a square")

# This class does NOT implement draw() – will cause an error if instantiated
class Triangle(Drawable):
    pass

# circle = Circle()   # OK
# square = Square()   # OK
# triangle = Triangle()  # TypeError: Can't instantiate abstract class Triangle with abstract methods draw

def render(shape):
    shape.draw()

render(Circle())   # Drawing a circle
render(Square())   # Drawing a square
```

```python
from abc import ABC, abstractmethod

class Payment(ABC):
    @abstractmethod
    def pay(self, amount):
        pass

class CreditCardPayment(Payment):
    def pay(self, amount):
        print(f"Paid ${amount} using Credit Card.")

class PayPalPayment(Payment):
    def pay(self, amount):
        print(f"Paid ${amount} using PayPal.")

def process_payment(payment_method, amount):
    payment_method.pay(amount)

# Test
cc = CreditCardPayment()
pp = PayPalPayment()
process_payment(cc, 100)
process_payment(pp, 100)

# This would raise an error because CashPayment doesn't implement pay()
# class CashPayment(Payment):
#     pass
# cash = CashPayment()   # TypeError!
```
