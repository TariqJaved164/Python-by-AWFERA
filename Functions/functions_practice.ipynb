## ğŸ” 3ï¸âƒ£ ATM Withdrawal Function
### ğŸ¯ Scenario:

A bank account has balance.
User wants to withdraw money.

Rules:

-   Cannot withdraw more than balance

-   Cannot withdraw negative amount

ğŸ“¥ Function takes:

-   balance

-   withdraw_amount

ğŸ“¤ Function returns:

-   new balance (or error message)

```python
def withdraw(balance, withdraw_amount):

    # input validation
    if withdraw_amount <= 0:
        return False , "Amount must be greater than 0."

    fee = withdraw_amount * 0.02
    total_deducted = withdraw_amount + fee

    if (total_deducted > balance):
        return False , "Insufficient balance"

    new_balance = balance - total_deducted

    return True ,{
        "Withdraw" : withdraw_amount ,
        "Fee" : fee ,
        "Total Deducted" : total_deducted ,
        "Remaining Balance" : new_balance
    }


success , result = withdraw(10000 , 1000)
if success:
    print("Withdrwal successfull")
    print(f"Withdraw :  {result["Withdraw"]}")
    print(f"Fee : {result["Fee"]}")
    print(f"Total Deducted : {result["Total Deducted"]}")
    print(f"Remaining Balance : {result["Remaining Balance"]}")

else:
    print(f"Error ! {result}")




```

```python
def calculate_delivery_charge(order_amount):

    if order_amount < 0:
        return "Order amount cannot be negative."

    if order_amount >= 3000:
        delivery_charge = 0

    elif order_amount >= 1000:
        delivery_charge = 200

    else:
        delivery_charge = 500

    return delivery_charge


# -------- Input Section --------
amount_input = input("Enter order amount: ")

if not amount_input.isdigit():
    print("âŒ Invalid input! Please enter a number.")
else:
    amount = int(amount_input)
    charge = calculate_delivery_charge(amount)
    print("Delivery Charge:", charge)
```

## ğŸ“š 5ï¸âƒ£ Library Fine Calculator
### ğŸ¯ Scenario:

Library charges:

-   10 rupees per late day

-   If days > 30 â†’ membership suspended (return special message)

ğŸ“¥ Function takes:

-   `late_days`

ğŸ“¤ Function returns:

-   `fine amount` OR suspension message

```python
def calculate_library_fine(late_days):

    if late_days < 0:
        return "Late days cannot be negative."

    if late_days == 0:
        return "No fine. Thank you for returning on time."

    if late_days > 30:
        return "Membership suspended due to excessive late return."

    fine = late_days * 10
    return fine


# -------- Input Section --------
days_input = input("Enter number of late days: ")

if not days_input.isdigit():
    print("âŒ Invalid input! Please enter a number.")
else:
    days = int(days_input)
    result = calculate_library_fine(days)
    print("Result:", result)
```

## ğŸ§¾ Problem 6 â€“ Electricity Bill Calculator
### ğŸ¯ Real-Life Scenario

We  are building a function for an electricity company.

The bill is calculated based on units consumed.

âš¡ Billing Rules

| Units Consumed | Rate Per Unit |
| -------------- | ------------- |
| 0 â€“ 100        | \$5 per unit  |
| 101 â€“ 300      | \$8 per unit  |
| Above 300      | \$10 per unit |


âš  Important:
This is slab-based billing, not flat billing.

That means:

-   First 100 units â†’ $5 each

-   Next 200 units â†’ $8 each

-   Remaining units â†’ $10 each

ğŸ§  Example

If units = 350

-   100 Ã— 5 = 500

-   200 Ã— 8 = 1600

-   50 Ã— 10 = 500

-   Total = 2600

```python
def calculate_bill(units):

    if units < 0:
        return False , "Units must be greater than or equal to 0."

    if units <= 100:
        energy_charge = units * 5

    elif units <= 300:
        energy_charge = (100 * 5) + ((units - 100) * 8)

    else:
        energy_charge = (100 * 5) + (200 * 8) + ((units - 300) * 10)


    tax = energy_charge * .05   # 5% tax
    total_bill = energy_charge + tax

    return True , {
        "units": units ,
        "energy_charge": energy_charge,
        "tax": tax,
        "total_bill": total_bill
    }

input_unit = input("Enter units consumed: ")

if not input_unit.isdigit():
    print("Invalid input. Please enter a number")
else:
    units = int(input_unit)

success , result = calculate_bill(units)

if success:
    print("\n--- Electricity Bill ---")
    print(f"Units Consumed: {result['units']}")
    print(f"Energy Charges: {result['energy_charge']}")
    print(f"Tax (5%): {result['tax']}")
    print(f"Total Bill: {result['total_bill']}")
else:
    print("Error! " , result)



```
